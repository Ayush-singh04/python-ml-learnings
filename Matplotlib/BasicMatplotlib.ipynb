```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,20,15,25]

plt.plot(x, y)
plt.show()
```

```python
# We gonna represent Bakery sales

x = ['Mon','Tue','Wed','Thur','Fri'] # Represent X Axis
y = [7,22,9,10,28] # Represent Y Axis

plt.plot(x, y)
plt.title("Bakery Sales This Week!")

plt.xlabel("Day of the Week")
plt.ylabel("Sales Per Day")

plt.show()
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

sales = [1293,8704,4110,5485,976,7706,2809,9598,5300,7000,6878,8930]

plt.plot(months, sales, color ='green', linestyle = '--', linewidth = 2, marker = 'o', label = '2024 Sales Data')
plt.xlabel("Months of 2024")
plt.ylabel("Sales Per Month")
plt.title("Monthly Sales Data Report") 

plt.legend(loc = "upper left" ,fontsize=7) # Helps to add Summrised title inside graph

plt.grid(color = "grey", linestyle = ":", linewidth = 1) # Adds Grid

# plt.xlim("Jun" , "Sep") # These methods Helps to Show specific data only
# plt.ylim(5000,10000) 
# plt.xticks([],[]) # Chanes the Complete series name
plt.show()
```

```python
# Bar chart/graph

product = ['Smart Phone','Head Phone','Laptop','Watch'] # Products
sales = [205500,22000,450000,90000] # Sales This Month 

plt.bar(product, sales, color = 'orange' , label = "Sales Aug-2025")
# plt.barh(product, sales, color = 'orange' , label = "Sales Aug-2025") # to represnt this bar chart horizontally
plt.xlabel("Products")
plt.ylabel("Sales in INR")
plt.title("Product Sales Comparison")
plt.legend()

plt.show()
```

```python
# pie chart/graph
Classes = ["10th-A","10th-B","10th-C","10th-D"]
Players = [3,1,2,5]

plt.pie(Players, labels = Classes, colors = ["blue","gold","lightgreen","purple"], autopct = "%1.1f%%")
plt.title("Players from each Class")
plt.show()
```

```python
# Histogram

Scores = [78, 45, 92, 61, 84, 33, 59, 98, 27, 69, 88, 12, 54, 73, 96, 41, 67, 25, 81, 38
]
plt.hist(Scores, bins=5, color = "green", edgecolor = "black") # bins = number of range we want to divide the data
plt.xlabel("Score Range")
plt.ylabel("Number of Students")
plt.title("Score Distribution of Students")
plt.show()
```

```python
#  Scatter Plot (Shows the Relation B/W two variables)

Schools = ["M.K. Memorial","Primary","Rajkiya Vid.","K.V","D.P.S","Goenka School"]
Passing_percentage = [60,65,72,80,90,99.9]

plt.scatter(Schools, Passing_percentage, color = "red", marker = "^", label = "Passing % of Schools")
plt.xlabel("Schools In District")
plt.ylabel("Passing % of Schools")
plt.legend(fontsize=7)
plt.grid(True)
plt.show()
```

```python
plt.scatter(["K.V","D.P.S","Goenka School"],[80,90,98], color = "red", marker = "^", label = "Passing % of School(In Delhi)")
plt.scatter(["K.V","D.P.S","Goenka School"],[85,96,100], color = "blue", marker = "^", label = "Passing % of School(In U.P)")

plt.xlabel("Schools of State")
plt.ylabel("Passing % of Schools")
plt.legend(fontsize=7)
plt.grid(True)
plt.show()
```

```python
#  Subplot (Creates More charts at single window/canvas)

x = ['Mon','Tue','Wed','Thur','Fri']
y = [7,22,9,10,28] 

plt.subplot(1, 2, 1) # 1st Row, 2nd Col, 1st subplot
plt.plot(x,y)
plt.title("Line Chart")

plt.subplot(1, 2, 2) # 1st Row, 2nd Col, 2nd subplot
plt.bar(x,y)
plt.title("Bar Chart")

plt.tight_layout()
plt.show()
```

```python
# Recomended Way for Subplot

fig, ax = plt.subplots(1, 2, figsize = (10,5)) # Imp Step

x = ['Mon','Tue','Wed','Thur','Fri']
y = [7,22,9,10,28] 

ax[0].plot(x,y , color = "black")
ax[0].set_title("Line Plot")

ax[1].bar(x,y , color = "green")
ax[1].set_title("Bar Chart")

fig.suptitle("Comparison of Line and Bar Charts ")

plt.tight_layout() # makes it look fit into the frame
plt.show()
```
