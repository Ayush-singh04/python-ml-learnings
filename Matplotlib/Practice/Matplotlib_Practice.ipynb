```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("netflix_dataset_raw.csv")

# df.head()
# df.isnull()
```

```python
# Cleaning Data
# df.info()
# df.columns.
df.dropna(subset=['show_id', 'type', 'title', 'director', 'cast', 'country', 'date_added',
       'release_year', 'rating', 'duration', 'listed_in', 'description',
       'imdb_score'])
```

```python
#  Checking the count of movies vs t.v shows

type_count = df["type"].value_counts()
plt.figure(figsize = (8,6))
plt.bar(type_count.index , type_count.values , color = ['orange','purple'])
plt.title("Number of Movies VS T.V Shows on Netflix")
plt.xlabel("Type of Content")
plt.ylabel("Quantity")

plt.tight_layout()
plt.show()
```
