```python
import matplotlib.pyplot as plt

#  Our Data
x = ['Mon','Tue','Wed','Thur','Fri']
y = [7,22,9,10,28] 

#  Creating Plot

plt.plot(x,y , color = "orange", marker = "o")
plt.title("Simple Line Plot")
plt.xlabel("Days of the Week")
plt.ylabel("Sold Cakes Per Day")

# Saving this figure
plt.savefig("Saved_linePlot.png" , dpi = 300, bbox_inches = "tight")
plt.show()
```
