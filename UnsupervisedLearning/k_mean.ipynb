```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# K means helps to form Clusters(group of similar data) as we are working on unsupervised learning(Data).
```

```python
#  Sample data

data = {
    'Customer': ["Riya","Aman","Rohit","Divyanshi","Prakhar","Astha",],
    'Age': [27,30,40,21,22,18],
    'Spending': [210,200,150,300,400,350]
}

df = pd.DataFrame(data)
```

```python
# Selecting Features for Clustering

X = df[["Age", "Spending"]]

# Creating KMeans model using sklearn

model = KMeans(n_clusters=2 , random_state=42, n_init=10)
```

```python
#  Created a Group where the model will store predicted Data.
df["Group"] = model.fit_predict(X)

plt.figure(figsize=(6,5))

# This step will find unique values in Group [0,1]
for group in df["Group"].unique(): 
    group_data = df[df["Group"] == group] # This step is called masking
    plt.scatter(group_data["Age"], group_data["Spending"], label=f'Group {group}')


plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.title("Customer Segements (K-Means)")
plt.legend()
plt.grid(True)
plt.show()

print(df)
```
